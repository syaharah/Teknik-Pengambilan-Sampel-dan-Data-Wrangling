```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import time
```

Scraping halaman 1- 5

```python
List_Elektronik = []

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "id-ID,id;q=0.9,en-US;q=0.8,en;q=0.7",
}
```

```python
for page in range(1, 6):  # halaman 1 sampai 5
    url = f"https://loginmegastore.com/search?query=kulkas&p=true&page={page}"
    response = requests.get(url, headers=headers, timeout=15)
    soup = BeautifulSoup(response.text, "html.parser")

    titles = soup.find_all("p", class_="mantine-focus-auto !leading-[120%] m_b6d8b162 mantine-Text-root")
    harga  = soup.find_all("p", class_="mantine-focus-auto !whitespace-nowrap m_b6d8b162 mantine-Text-root")
    gambar = soup.find_all("img", class_="m_9e117634 mantine-Image-root")

    for i in range(min(len(titles), len(harga), len(gambar))):
        item = {
            "Title" : titles[i].get_text(strip=True),
            "Harga" : harga[i].get_text(strip=True),
            "Gambar": gambar[i].get("src", "")
        }
        List_Elektronik.append(item)

    print(f"Halaman {page}: {len(titles)} produk ditemukan")
    time.sleep(1)  # jeda agar tidak di-block server
```

```python
with open("List_Elektronik.json", "w", encoding="utf-8") as f:
    json.dump(List_Elektronik, f, ensure_ascii=False, indent=4)

print(f"\nTotal data: {len(List_Elektronik)} produk")
print("File List_Elektronik.json berhasil disimpan!")
```

Berikan nama hasil scraping bernama List_Elektronik, ubah format file menjadi JSON

```python
with open("List_Elektronik.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

Ubah format data menjadi csv menggunakan prinsip ETL.

```python
df = pd.DataFrame(data)

# Bersihkan kolom Harga (hapus "Rp " dan titik ribuan)
df["Harga_Clean"] = (
    df["Harga"]
    .str.replace("Rp ", "", regex=False)
    .str.replace(".", "", regex=False)
    .astype(int)
)

df["Sumber_Halaman"] = "loginmegastore.com"

print("Info DataFrame:")
print(df.info())
print("\nPreview 5 data:")
print(df.head(5))
```

```python
df.to_csv("List_Elektronik.csv", index=False, encoding="utf-8")
print("\nFile List_Elektronik.csv berhasil disimpan!")
```
